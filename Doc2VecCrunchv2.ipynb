```python
def read_corpus(fname, tokens_only=False):
    with smart_open.open(fname, encoding="iso-8859-1") as f:
        for i, line in enumerate(f):
            tokens = gensim.utils.simple_preprocess(line)
            if tokens_only:
                yield tokens
            else:
                # For training data, add tags
                yield gensim.models.doc2vec.TaggedDocument(tokens, [i])

def find_tokens(fname, token1):
    ia=[]
    with smart_open.open(fname, encoding="iso-8859-1") as f:
        for i, line in enumerate(f):
            if token1 in line:
                print("tinyML here:",i)
                ia.append(i)
                
    ianp=np.array(ia)
    return ianp


#train_corpus = list(read_corpus(lee_train_file))
#test_corpus = list(read_corpus(lee_test_file, tokens_only=True))
train_corpus=list(read_corpus(startup_train_file))
token1="federated"
ianp1=find_tokens(startup_train_file,token1)
print("ianp back",ianp1,ianp1.shape[0])
test_corpus=list(read_corpus(startup_test_file, tokens_only=True))
ianp2=find_tokens(startup_test_file,token1)
print("ianp back",ianp2)
print("len train corpus",len(train_corpus))
# print(test_corpus[:2])
```

```python
dim_vector=25
model = gensim.models.doc2vec.Doc2Vec(dm=1, vector_size=dim_vector, min_count=2, epochs=200)
model.build_vocab(train_corpus)
# print(f"Word 'penalty' appeared {model.wv.get_vecattr('penalty', 'count')} times in the training corpus.")
model.train(train_corpus, total_examples=model.corpus_count, epochs=model.epochs)
# vector = model.infer_vector(['only', 'you', 'can', 'prevent', 'forest', 'fires'])
# print(vector)
```
